# First Order Motion Model for Image Animation
Aliaksandr Siarohin, Stéphane Lathuilière, Sergey Tulyakov, Elisa Ricci and Nicu Sebe did an amazing, simply phenomenal work called the "First Order Motion Model for Image Animation". Their technology allows you to read facial expressions and movements from video and supplement the static image on the basis of this, predicting what parts that are not in the original image will look like.
```
import imageio
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from skimage.transform import resize
from IPython.display import HTML

image = imageio.imread('/content/gdrive/My Drive/Deep/dicaprio2.jpg')
video = imageio.mimread('/content/gdrive/My Drive/Deep/10.mp4')

image = resize(image, (256,256))[..., :3]
video = [resize(frame,(256,256))[..., :3] for frame in video]

def display(source, driving, generated = None):
  fig = plt.figure(facecolor = 'black',figsize = (8 + 4 * (generated is not None), 6))
  plt.style.use('dark_background')
  #plt.text(1,-18,s = 'Original image',fontsize = 18,  ha='left',
         #va='top', wrap=True)
  if generated is not None:
    plt.title('Source video:',fontsize = 18)
    plt.title('Original image:',fontsize = 18, loc='left')
    plt.title('Result:',fontsize = 18, loc='right')

  ims = []
  for i in range(len(driving)):
    cols = [source]
    cols.append(driving[i])
    if generated is not None:
      cols.append(generated[i])
    im = plt.imshow(np.concatenate(cols, axis = 1), animated = True)
    plt.axis('off')
    
    ims.append([im]) 
  
  ani = animation.ArtistAnimation(fig, ims, interval = 50, repeat_delay = 0)
  plt.close()
  return ani

HTML(display(image, video).to_html5_video())
```
